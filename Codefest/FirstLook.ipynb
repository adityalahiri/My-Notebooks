```python
#importing datasets. 3 training files
problem_df = pd.read_csv("problem_data.csv")
user_df = pd.read_csv("user_data.csv")
submissions_df = pd.read_csv("train_submissions.csv")
```

```python
# attempting to merge user_df on basis of user_id into submissions_df

sub_user_merge = pd.merge(submissions_df, user_df, on='user_id', how='outer')
```

```python
sub_user_merge.head()
```

```python
# attempting to merge sub_user_merge with problem_df on basis of prolem_id

final_merge = pd.merge(sub_user_merge, problem_df, on='problem_id', how='outer')
```

```python
# removing if nan in user_id or problem_id

final_merge = final_merge[pd.notnull(final_merge['user_id'])]
final_merge = final_merge[pd.notnull(final_merge['problem_id'])]
```

```python
# saving this datframe

final_merge.to_csv("merged_dataset")
```
